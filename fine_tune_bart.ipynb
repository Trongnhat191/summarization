```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# model_name = 'google/flan-t5-large'
model_name = 'facebook/bart-large'
model = AutoModelForSeq2SeqLM.from_pretrained(model_name,
                                              cache_dir = cache_dir,
                                              torch_dtype = torch.float16).to(device)
tokenizer = AutoTokenizer.from_pretrained(model_name,
                                          cache_dir = cache_dir)
```

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=8,                 # rank c·ªßa ma tr·∫≠n th·∫•p (c√†ng cao ‚Üí c√†ng ch√≠nh x√°c, nh∆∞ng t·ªën VRAM h∆°n)
    lora_alpha=16,       # h·ªá s·ªë ƒëi·ªÅu ch·ªânh
    target_modules=["q_proj", "v_proj"],  # ch√®n LoRA v√†o c√°c l·ªõp attention
    lora_dropout=0.05,   # tr√°nh overfitting
    bias="none",
    task_type="SEQ_2_SEQ_LM"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

### Dataset

```python
from datasets import load_dataset

data = load_dataset('OpenHust/vietnamese-summarization')
```

```python
train_data = data['train']
train_data
```

```python
filtered_dataset = train_data.filter(lambda x: len(x["Document"].split()) < 512)
print(len(filtered_dataset))
```

```python
filtered_dataset = filtered_dataset.select(range(20000))
```

```python
filtered_dataset
```

```python
filtered_dataset
```

```python
max_input_length = 512
max_target_length = 256

def preprocess_function(examples):
    inputs = examples["Document"]
    # print(inputs)
    targets = examples["Summary"]
    # print(targets)
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)
    labels = tokenizer(targets, max_length=max_target_length, truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_datasets = filtered_dataset.map(preprocess_function, batched=True)
# train_test_split tokenized_datasets
tokenized_train = tokenized_datasets.train_test_split(test_size=0.1, shuffle=True)
tokenized_eval = tokenized_train['test']
print(len(tokenized_train['train']))
print(len(tokenized_eval))
```

```python
print(len(tokenized_train['train']))
print(len(tokenized_eval))
```

```python
tokenized_eval
```

```python
from transformers import DataCollatorForSeq2Seq, TrainingArguments, Trainer

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

training_args = TrainingArguments(
    output_dir="lora_large_bart_fb",
    num_train_epochs=6,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,  # n·∫øu GPU nh·ªè
    save_steps = 200,
    logging_steps=200,
    learning_rate=2e-4,             # LR cao h∆°n fine-tune full
    per_device_eval_batch_size=4,
    logging_dir = './logs',
    # logging_strategy='steps',
    # eval_strategy="epoch",
    # save_strategy="epoch",
    fp16=True,
    # remove_unused_columns=False,
    # load_best_model_at_end=True,
    # save_total_limit=2,
    # report_to="none"
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train['train'],
    eval_dataset=tokenized_eval,
    processing_class=tokenizer,
    data_collator=data_collator,
)

trainer.train()
```

# IV. Evaluation after fine tuning

```python
# test_infer_eval_samples.py

import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from datasets import load_dataset


# Load model and tokenizer
model_dir = "lora_fb_bart_large_adapter_20K"
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = AutoModelForSeq2SeqLM.from_pretrained(model_dir).to(device)
tokenizer = AutoTokenizer.from_pretrained('facebook/bart-large')
print('done model')

# Load eval dataset
data = load_dataset('OpenHust/vietnamese-summarization')
eval_data = data['train'].filter(lambda x: len(x["Document"].split()) < 512).select(range(20000))
tokenized_eval = eval_data.train_test_split(test_size=0.1, shuffle=True)['test']

```

```python
# Infer 5 samples
for i in range(5):
    print('--------------------------')
    text = tokenized_eval[i]['Document']
    print(text)
    text = 'T√≥m t·∫Øt : ' + text.replace('_', ' ')

    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512).input_ids.to(device)
    model.eval()
    with torch.no_grad():
        outputs = model.generate(inputs, max_new_tokens=256, do_sample=True)
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"Sample {i+1} summary:", summary)
    assert isinstance(summary, str)
    assert len(summary.strip()) > 0
```
